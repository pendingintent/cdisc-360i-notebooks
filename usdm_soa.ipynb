# Create a Schedule of Acitivities from USDM JSON #

```python
# import required libraries
import pandas as pd
import json
```

```python
# read the USDM JSON for the LZT study
in_file = "data/usdm.json"
try:
    with open(in_file, 'r') as f:
        usdm = json.load(f)
except FileNotFoundError:
    print('The file {} was not found...'.format(in_file))
```

```python
# Return the STUDYID
studyId = usdm['study']['versions'][0]['studyIdentifiers'][0]['text']
```

```python
# Create the required study dictionary lists
activities = usdm['study']['versions'][0]['studyDesigns'][0]['activities']
encounters = usdm['study']['versions'][0]['studyDesigns'][0]['encounters']
timelines = usdm['study']['versions'][0]['studyDesigns'][0]['scheduleTimelines']
```

```python
#
timelinesDF = pd.DataFrame(timelines, columns=['name','description','mainTimeline','entrytId','timings','instances'])
mainTimelineDF = timelinesDF[timelinesDF['mainTimeline']]
mainTimelineDF
```

```python
# Create the column headers for the SoA using the encounter label
encountersDF = pd.DataFrame(encounters)
#encountersDF = encountersDF.T
encountersDF.head()
```

```python
# Create the pandas dataframe
activitiesDF = pd.DataFrame(activities, columns = ['id','label','biomedicalConceptIds','bcCategoryIds','bcSurrogateIds'])
activitiesDF
```

```python
df = pd.DataFrame(data=activitiesDF['label'], columns = encountersDF)
df
```
